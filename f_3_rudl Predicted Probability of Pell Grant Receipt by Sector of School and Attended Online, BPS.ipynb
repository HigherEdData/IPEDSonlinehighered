### Note:
The following code marks "good" for analysis if everonline10 is a dichotomous measure restricted to repondents in our sample- those whose first institution attended was a four-year institution.
They are coded as 1 if their entire certificate or degree program at their first, last, or both first and last institution was online. They are coded as 0 if attendance at either four-year institution was never online. 
They are missing if their first institution attended was NOT a four-year institution.

```stata
qui mark good if everonline10!=.
```

```stata
rename pell Pell
```

```stata
qui {

logistic Pell i.everonline10 i.everforprofit10 i.everprivate10 ///
i.highestdegree i.HBCU i.everonline10##i.everforprofit10, cluster (cluster9)

margins i.everonline10 if everforprofit10==1, atmeans 

marginsplot, name(forprof, replace) recast(bar) scheme(plotplain) ///
ytitle(Probability of Pell Grant Receipt) title (For Profit, size(large)) plotopts(barw(.8))

margins i.everonline10 if everforprofit10==0, atmeans 

marginsplot, name(notforprof, replace) recast(bar) scheme(plotplain) ///
ytitle(Probability of Pell Grant Receipt) title (Not For Profit, size(large)) plotopts(barw(.8)) ///
yscale(range(0 1))
}
```

```stata
graph combine notforprof forprof, row(1) ycommon name(pell , replace)
```
